**Q8. How would you define the superclasses of a class?**

**Solution:-** A superclass is the class from which many subclasses are created. The subclasses inherit the characteristics of a superclass. The superclass is also known as the parent class or base class. The **super() function** is used to give access to methods and properties of a **parent or sibling** class to subclasses. The **super() function** returns an object that represents the **parent class.**

```python
class Parent:
    
    def __init__(self, txt):
        self.message = txt

    def printmessage(self):
         print(self.message)

class Child(Parent):
    
    def __init__(self, txt):
        super().__init__(txt)

x = Child("Hello, and welcome!")

x.printmessage()
```
